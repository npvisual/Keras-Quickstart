### 2D Convolution : 1st layer

````python
model = Sequential()

# Original 1st layer in the model from the tutorial was replaced by the next
# 3 lines, which are equivalent. They just break out the activation layer so
# it can be used by the visualization function later on.
#
# model.add(Convolution2D(32, 3, 3, activation='relu', input_shape=(1,28,28)))

# Layer #1
model.add(Convolution2D(32, 3, 3, input_shape=(1,28,28)))
# Layer #2
convout1 = Activation('relu')
model.add(convout1)
````

However... As opposed to the `keras.io` documentation, which provides no contextual information on the parameter values of each layer, the tutorial here provides a lot of good information, which I will just paste below because it's important :

<div class="alert alert-block alert-info">
   The input shape parameter should be the shape of 1 sample. In this case, it's the same (1, 28, 28) that corresponds to  the (depth, width, height) of each digit image.<br>But what do the first 3 parameters represent? It seems that the size of the `input_shape` parameter is pretty well explained. Essentially, since this is the first layer in the model, it has to fit whatever we're feeding to it. I.e. gray-scale (1 dimension) pictures which are 28x28 pixels "large". So far so good...

The `Conv2D` layer is [pretty well documented](https://keras.io/layers/convolutional/#conv2d), compared to others, to be fair, but it's still hard to grasp what the __filters__ do and what the __convolution kernel__ size means... After all, if your best-friend were to tell you :

<div class="alert alert-block alert-info">This layer creates a convolution kernel that is convolved with the layer input to produce a tensor of outputs.</div>

you'd strongly consider finding a new best-friend...

Luckily, there are a few articles out there which provide some good explanations (supported by some nice graphics) of what a convolution kernel is :
* [refresher on discrete convolution](http://deeplearning.net/software/theano_versions/dev/tutorial/conv_arithmetic.html#refresher-discrete-convolutions) -- really cool article.
* [understanding convolutional neural networks](http://www.wildml.com/2015/11/understanding-convolutional-neural-networks-for-nlp/)

So the __kernel__ is that "shaded" area that goes over the matrix and combines (convolves) with it to produce a smaller matrix -- in this case a new `26x26` matrix.

The first parameter (`32`) defines how many of those __kernels__ get applied to the input (the image).

__"Why `32`"__ ? Don't know, but the reason they are called __filters__ is because they each detect a specific "feature" of the image. So each one of those `3x3` matrix has a very specific set of values that's supposed to trigger or isolate a "feature" of the image. As a result there are __32 output nodes__ when this layer is done processing an image.

Now in the "Refresher on discrete convolution", Denny Britz makes a pretty powerful statement :

<div class="alert alert-block alert-info">
a CNN automatically learns the values of its filters based on the task you want to perform
</div>

My thinking was : "well, we only have positive values in the input matrix (values between 0 and 1) and therefore all values in the output matrices should be positive". Right ?

__Wrong !__ I had forgotten about the values of the __kernels__ / __filters__. Those could be negative. For example the ["Edge Detect" filter](https://docs.gimp.org/en/plug-in-convmatrix.html) has a `-4` value in its center, which could lead to values in the matrix being negative. There is also a nasty little side-effect of the convolution (which we will see later when visualizing the results) : the background switches from white to black, changing the fundamental trait of the input set we fed to the model.

Let's look at the [Rectified Linear Unit](https://en.wikipedia.org/wiki/Rectifier_(neural_networks) (ReLU for short) function for a sec -- the code provided in Wikipedia by Dan Stowell was actually quite interesting so I included it in this notebook, just for fun !

````python
# coding=UTF8
#
# python code to plot nonlinearity functions
# written 2015 by Dan Stowell, dedicated to the public domain

plt.rcParams.update({'font.size': 6})

nlfuns = {
        'Rectifier':     ('b', '', lambda x: np.maximum(0, x)),
        'Softplus':      ('g', '', lambda x: np.log(1 + np.exp( 1 * x))/ 1),
#       'Softplus c=10': ('r', '', lambda x: np.log(1 + np.exp(10 * x))/10),
#       'Exponential':   ('c', '', lambda x: np.exp(x)),
}

evalpoints = np.linspace(-4, 4, 51)

params = {
   'axes.labelsize': 10,
   'font.size': 10,
   'legend.fontsize': 10,
   'xtick.labelsize': 8,
   'ytick.labelsize': 8,
   'text.usetex': False,
   'figure.figsize': [5.5, 4],
   'lines.markersize': 4,
}
plt.rcParams.update(params)


plt.figure(frameon=False)
plt.axes(frameon=0)
plt.axvline(0, color=[0.6]*3)
plt.axhline(0, color=[0.6]*3)
for nlname, (color, marker, nlfun) in nlfuns.items():
        plt.plot(evalpoints, map(nlfun, evalpoints), label=nlname, color=color, marker=marker)
plt.title('Nonlinearities')
plt.xlim(-4, 4)
plt.ylim(-0.1, 4.5)
plt.legend(loc='upper left', frameon=False)
plt.xlabel('x')
plt.ylabel(u'σ(x)')
plt.show()
#plt.savefig('Rectifier_and_softplus_functions.svg')
plt.close()
````

### Next layers

Let's look at the next layers and the shape of their output.

````python
# Same comment here : the original layer from the tutorial is commented out
# and the activation function is extracted into its own layer for the visualization.
#
# model.add(Convolution2D(32, 3, 3, activation='relu'))

# Layer #3
model.add(Convolution2D(32, 3, 3))
print model.output_shape
# Layer #4
convout2 = Activation('relu')
model.add(convout2)
print model.output_shape
# Layer #5
model.add(MaxPooling2D(pool_size=(2,2)))
print model.output_shape
# Layer #6
model.add(Dropout(0.25))
print model.output_shape
````

The next layer added to the model is the exact [same Conv2D layer that was used in the first layer](#2D-Convolution-:-1st-layer). The only difference is that the input parameters are not present as they are gathered by the model automatically from the previous layer. That's pretty straightforward so far.

```
    # Layer #1
    model.add(Convolution2D(32, 3, 3, input_shape=(1,28,28)))
    # Layer #3
    model.add(Convolution2D(32, 3, 3))
```

So same number (`32`) of `3x3` convolution kernels and same activation function (ReLU). This results in 32 nodes of `24x24` pixels across the 60k samples.

Again, how they came up with `32` as the number of kernels (i.e. "features") is still a mystery to me.

We'll see later that the [results of the 2nd convolution](#Visualizing-the-2nd-convolution), especially [after going through the activation function](#Activation-layer-after-2nd-convolution), show some pretty distincts features.

The 5th layer uses [MaxPooling2D](https://keras.io/layers/pooling/#maxpooling2d) with a `pool_size` of (2,2) which is clearly described in the Keras documentation :

<div class="alert alert-block alert-info">(2, 2) will halve the input in both spatial dimension.</div>

A [max pooling function](https://en.wikipedia.org/wiki/Convolutional_neural_network#Pooling_layer) with a pool size of 2 along both width and height, with an identical `stride` (none is specified here, so it defaults to the pool size) will shrink the input matrice in 2, resulting in 32 matrices of 12x12 pixels. The `max` function keeps the notion that a "scribble" is present in the area but looses the space referential (data loss, but good enough approximation).

The [results of this layer](#Onto-the-max-pooling-layer) clearly demonstrate the loss of of information : without knowing what the original digit was, it's hard to guess based on images from this layer. However they clearly highlight __separate and distinct features, like horizontal or diagonal lines__.

The 6th layer, `Dropout`, prevents the model from becoming too biased toward the training dataset (what folks call "overfitting"). The tutorial points to a [great set of answers](https://www.quora.com/How-does-the-dropout-method-work-in-deep-learning-And-why-is-it-claimed-to-be-an-effective-trick-to-improve-your-network) to explain what this layer does. The `rate` parameter of `0.25` indicates that a quarter of all the input units (chosen randomly) to this layer will be silenced / tuned out during each iteration. Again, __"why 1/4 now and 1/2 later (in layer #9)"__ noone seems interested in providing any answers or context.

It could be [empirical evidence](https://en.wikipedia.org/wiki/Empirical_evidence) or there could be a well-known receipe to avoid overfitting... I just don't know.

But the intent seems straightforward to me.

_Note : there were no [noticeable visual differences](#Visualizing-drop-out) between layers `#5` and `#6` and this had to do with being in testing vs. training mode._

### Final layers ?

Ok, so now that we've reduced the matrix size (12x12) and prevented over-fitting, what's next ?

````python
# Layer #7
model.add(Flatten())
print model.output_shape
# Layer #8
model.add(Dense(128, activation='relu'))
print model.output_shape
# Layer #9
model.add(Dropout(0.5))
print model.output_shape

# Original layer as provided in the tutorial
# model.add(Dense(nb_classes, activation='softmax'))

# Layer #10
model.add(Dense(nb_classes))
# Layer #11
model.add(Activation('softmax'))
print model.output_shape
````

#### Fully-connected layer

Alright. So the __`Flatten` layer__ might be easy, but requires, nevertheless, an explanation : we're transforming a 3-dimensional array (32x12x12) into a vector. The size of the vector is 4608 (=32x12x12).

So why do we do this ? Well, it looks like we're moving from a convolutional layer to a fully-connected layer, which requires 1 dimensional arrays (vectors). Every node in a layer is connected to every node in the next layer ; and every connection has a "weight".

Now, my understanding of `Dense` (pun intended), in Keras, is that it doesn't require the input to be flattened before, as it will [perform that operation itself if the input isn't a vector](https://keras.io/layers/core/#dense) (see the Note). So maybe that's something that I will need to validate down the road...

Anyhow, __the next layer, `Dense`,__ will produce a vector whose size is 128 from the 4608-long input vector, using the ReLU activation function on each value of the output vector. So my guess here is that we're still trying to get rid of the outliers.

But __"why 128?"__, you might say... No idea, at least for now.

_Note : I kept the compound form, where the activation function is a parameter of the `Dense` layer, just to illustrate that both can be used._

#### Avoiding overfitting

Next up the model is trying to prevent some more overfitting by using __`Dropout`__ and kicking half (`rate=0.5`) of the neurons out of the equation during each cycle. The shape of the model remains the same (128).

#### On more fully-connected layer to make it home !

The last step (layers #10 & #11) is yet another fully-connected neural net layer, using __`Dense`__. But THIS TIME the model's size is finally reduced to the number of categories, i.e. `10`. This is the last layer in the model and it makes perfect sense that it would produce 10 final neurons / nodes. Each one of those neurons should provide an indication of which digit the image fed to the model is most likely to match.

That would, however, mean that the values in each category correspond to some __probability__ of the input image being a specific digit. Unfortunately, after this last `Dense` layer those values are nowhere near normalized.

Enters the __`softmax` function__ (last layer) ! I'll give you a few minutes to [read about it](https://github.com/Kulbear/deep-learning-nano-foundation/wiki/ReLU-and-Softmax-Activation-Functions)... Good ? So just to recap, softmax will take all those values and "rearrange" them so that the sum of them all equals 1.

Now we have a probability result of the image belonging to any of the 10 categories.

I think `softmax` is now officially my __favorite activation function__.

### Training the model

Well... here we are ! So... __fair warning__ : it takes about 2 min per epoch (cycle) on my Macbook Pro to train the model on the MNIST dataset. Not bad, but considering that we loop 12 times, you can go out for a quick run and come back in 30 min (plus that will offset all that caffeine you just drank).

_Note : if you look at the [tutorial](https://elitedatascience.com/keras-tutorial-deep-learning-in-python#step-9) you should start seeing a big discrepancy in the code below. This is mostly due to the addition of the visualization code. I added it because I REALLY, REALLY wanted to see the resulting images as they are going through the model. Nuff said, put your sneakers on..._

````python
# Original training command from the tutorial. I adapted the code used
# by Julien Rebetez to match the batch size of the tutorial
#
# model.fit(X_train, Y_train, 
#           batch_size=32, nb_epoch=10, verbose=1)

WEIGHTS_FNAME = 'mnist_cnn_weights.hdf'
if True and os.path.exists(WEIGHTS_FNAME):
    # Just change the True to false to force re-training
    print('Loading existing weights')
    model.load_weights(WEIGHTS_FNAME)
else:
    batch_size = 32
    nb_epoch = 12
    model.fit(X_train, Y_train, batch_size=batch_size, nb_epoch=nb_epoch,
              verbose=1, validation_data=(X_test, Y_test))
    model.save_weights(WEIGHTS_FNAME)
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test score:', score[0])
print('Test accuracy:', score[1])
````

So we've trained our model ; the weights are saved so we can reuse them later (very useful on large training sets !) ; we now have a nice little output around each epoch :
* the time taken (in seconds),
* the loss score (loss) based on the categorical cross-entropy function,
* the accuracy (acc) and...
* wait... what !?!?! What are those : val_loss & val_acc ?

Well those are the loss and accuracy, respectively, for the __validation set__ that we used. Remember that we used a training set (X_train with 60k images) and a validation set (X_test with 10k images). So Keras computes loss and accuracy for each one of those sets.

There are numerous discussions on [which ones to use](https://stackoverflow.com/questions/34702041/how-to-tell-which-keras-model-is-better) and why those values are different ; difference in dataset (which makes the most sense to me), difference around when it's computed, etc. I think this will make a nice little topic of research for another Jupyter notebook...

However you can see that the `Test score` and `Test accuracy` provided once the model is done training are the ones given by the last run for the validation set (rounded to closest decimal).

#### Prediction on the entire validation set

So now let's run the prediction for the entire validation set. The `argmax` function is then used to convert the category-based matrices of predictions to a vector of indices. This sounds a little complex, but essentially what it does is to convert the `0/1` positions into their actual digit counterparts... We're basically transforming the category matrix back to a list of values.

If you remember the step we took with the [data cleanup](#Cleaning-the-data), this is the reverse operation, so we can infer, from the category matrix, what the actual digit is.

````python
Y_pred = model.predict(X_test)
# Convert one-hot to index
y_pred = np.argmax(Y_pred, axis=1)
print(y_pred[0:5])
````

This shows the first 5 digits in the validation set : `7, 2, 1, 0` and `4`, which is exactly what we had before when we printed the first five labels of `y_test`.

#### Classification report

[Scikit](http://scikit-learn.org/stable/) provides a [classification report](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html#sklearn.metrics.classification_report) which outputs a set of metrics, for each category, i.e. digit, based on the results of the prediction on the validation set. I encourage you to read the [explanation](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_recall_fscore_support.html#sklearn.metrics.precision_recall_fscore_support) of what each column means.

Here's a summary (definitions provided by the SciKit user guide in bold, and my comments in italic) :
* __precision : the precision is intuitively the ability of the classifier not to label as positive a sample that is negative.__  _Sensitivity to false positive ?_
* __recall : the recall is intuitively the ability of the classifier to find all the positive samples.__ _Sensitivity to false negative ?_
* __f1-score : the F-beta score can be interpreted as a weighted harmonic mean of the precision and recall.__ _Hehe... good luck with that... More seriously, this should be the sensitivity ratio of false positives to false negatives._
* __support : the support is the number of occurrences of each class.__ _This is actually really useful as it shows whether or not the validation set is biased toward a particular category ; in this case it looks like each digit was represented pretty fairly in the validation set (except maybe for `5`)._

````python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
````

The classification report looks really good. There might be a very slight tendency of the model to not recognize the digit `9` (20 out of 1009) but all things considered, that's pretty good !

### Visualization

Now we're into the part that I got really excited about midway into working on this notebook : visualizing the images as they go through the layers and the filters. So let's get to it.

From what I understand from Julien's notebook, below we're defining a callback function (`convout1_f`) for the first and second convolution layers.

_Note : I updated the code to use `K.function` since I had some issues with the original code when trying to display the `Dropout` layer further down._

````python
from keras import backend as K
convout1_f = K.function([model.layers[0].input], convout1.output)
convout2_f = K.function([model.layers[0].input], convout2.output)
````

Next we just define a wrapper around `pl.imshow` in order to display a legend for the gray-scale gradient. And we display our first image, the digit `7`, with this new function called `nice_imshow`.

````python
# utility functions
from mpl_toolkits.axes_grid1 import make_axes_locatable

def nice_imshow(ax, data, vmin=None, vmax=None, cmap=None):
    """Wrapper around pl.imshow"""
    if cmap is None:
        cmap = cm.jet
    if vmin is None:
        vmin = data.min()
    if vmax is None:
        vmax = data.max()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    im = ax.imshow(data, vmin=vmin, vmax=vmax, interpolation='nearest', cmap=cmap)
    pl.colorbar(im, cax=cax)
````

````python
i = 0

# Visualize the input of the first layer of convolutions
X = X_test[i:i+1]

pl.figure()
pl.title('input')
nice_imshow(pl.gca(), np.squeeze(X), vmin=0, vmax=1, cmap=cm.binary)
````

Below is yet another helper function that will display a mosaic of images, which we will later pass to `nice_imshow` to display the output of certain layers in our neural network.

````python
import numpy.ma as ma
def make_mosaic(imgs, nrows, ncols, border=1):
    """
    Given a set of images with all the same shape, makes a
    mosaic with nrows and ncols
    """
    nimgs = imgs.shape[0]
    imshape = imgs.shape[1:]
    
    mosaic = ma.masked_all((nrows * imshape[0] + (nrows - 1) * border,
                            ncols * imshape[1] + (ncols - 1) * border),
                            dtype=np.float32)
    
    paddedh = imshape[0] + border
    paddedw = imshape[1] + border
    for i in xrange(nimgs):
        row = int(np.floor(i / ncols))
        col = i % ncols
        
        mosaic[row * paddedh:row * paddedh + imshape[0],
               col * paddedw:col * paddedw + imshape[1]] = imgs[i]
    return mosaic
````

#### Visualizing the weights of the 1st convolution

The 'weights' of the 1st convolution layer are the __filters__ that we talked about earlier : i.e. those `3x3` matrices that will move over the input data and convolve with them to start showing a particular feature of an image. So each one of those `32` filters has a specific purpose.

Those values were not normalized between `0` and `1`, but you can see with the scale on the right side what level of 'gray' they correspond to.

````python
# Visualize weights for convout1
W = model.layers[0].W.get_value(borrow=True)
W = np.squeeze(W)
print("W shape : ", W.shape)
print("W first weights : ", W[0])

pl.figure(figsize=(15, 15))
pl.title('conv1 weights')
nice_imshow(pl.gca(), make_mosaic(W, 6, 6), cmap=cm.binary)
````

#### Visualizing the 1st convolution

Let's now take a look at the transformed images after the 1st convolution layer, but BEFORE the activation layer.

_Note : this seemed important to me, in order to show what the activation layer really does. ReLU isn't that straightforward on its own, so being able to visualize the result of the activation function __later__ made more sense to me, rather than directly showing the results past the activation layer._

````python
# Visualize max pooling results for the 5th layer
convout1_0_f = K.function([model.layers[0].input], model.layers[0].output)
C1_0 = convout1_0_f([X])[0]
C1_0 = np.squeeze(C1_0)
print("C1_0 shape : ", C1_0.shape)

pl.figure(figsize=(15, 15))
pl.suptitle('convout1_0')
nice_imshow(pl.gca(), make_mosaic(C1_0, 6, 6), cmap=cm.binary)
````

As you can see above, there are already some pretty stricking features which are starting to emerge from the 1st convolution layer.

For example, the images on (row 2, column 2) and (row 4, column 5) seem to highlight the "diagonal" feature of the digit `7`.

Now the other interesting feature about those images is that they all have dark backgrounds now. The next layer (activation function) seems to take care of reverting the background to white. It also highlights even further those 'salient' points that seem to emerge.

To be fair this was a little counter-intuitive to me at the beginning, until I looked further at the scale on the right side of the mosaic. ReLU isn't supposed to 'activate' until a fairly dark level of gray pixels (at `0`). And, as you will see in the next mosaic, only 5 of those images retain a somewhat grayish background.

Therefore, to be retained as a feature of a digit, the pixels had to be fairly dark.

#### Visualizing the results of the activation function after 1st convolution

The ReLU activation function comes into full force here and will literally wipe out the 'noise' that is left in the images above. This results in a set which looks a lot more like salient features rather than a proper digit.

But you can still guess at what number the image represents. Therefore the result of the first `convolution + activation` is probably not enough to identify individual features of digit categories.

````python
# Visualize convolution result (after activation)
C1 = convout1_f([X])[0]
C1 = np.squeeze(C1)
print("C1 shape : ", C1.shape)

pl.figure(figsize=(15, 15))
pl.suptitle('convout1')
nice_imshow(pl.gca(), make_mosaic(C1, 6, 6), cmap=cm.binary)
````

#### Visualizing the 2nd convolution

The 2nd convolution results in smaller images of `24x24` pixels. While we can still easily guess what digit it is, the images are blurry and the features more extreme : the salient lines are either much darker (img r2, col 2) or much lighter (img r6, col1) compared to the result of the 1st convolution. 

````python
# Visualize convolution results for the 2n convolution layer (before activation)
convout2_0_f = K.function([model.layers[0].input], model.layers[2].output)
C2_0 = convout2_0_f([X])
C2_0 = np.squeeze(C2_0)
print("C2_0 shape : ", C2_0.shape)

pl.figure(figsize=(15, 15))
pl.suptitle('convout2_0')
nice_imshow(pl.gca(), make_mosaic(C2_0, 6, 6), cmap=cm.binary)
````

#### Activation layer after 2nd convolution

The same comments apply for this 2nd activation layer as for the 1st one.

But now it's __impossible to guess which digit we're looking at__, unless we look at the entire collection of those `32` images. And this is what convolutional neural nets are really good about : __extracting individual features from complex objects__ (granted, digits aren't that complicated !).

````python
# Visualize convolution result (after activation) for the 2n convolution layer
C2 = convout2_f([X])
C2 = np.squeeze(C2)
print("C2 shape : ", C2.shape)

pl.figure(figsize=(15, 15))
pl.suptitle('convout2')
nice_imshow(pl.gca(), make_mosaic(C2, 6, 6), cmap=cm.binary)
````

Imo, this is really where we see the neurons that fire on specific digits and highlight a set of features.

For example img r6, col1 that we refered to earlier (overall lighter shades of gray), is almost a muted neuron for the digit `7` (only one pixel). But if we look at the eighth image -- which depicts a `9`, with a weird shape ! -- then this neuron is fired up by the slanted leg of the `9`.

````python
i = 7 # digit 9
Z = X_test[i:i+1]
pl.figure()
pl.title('input')
nice_imshow(pl.gca(), np.squeeze(Z), vmin=0, vmax=1, cmap=cm.binary)
# Visualize convolution result (after activation) for the 2n convolution layer for the digit 9
C2 = convout2_f([Z])
C2 = np.squeeze(C2)
pl.figure(figsize=(15, 15))
pl.suptitle('convout2 for nine')
nice_imshow(pl.gca(), make_mosaic(C2, 6, 6), cmap=cm.binary)
````

#### Onto the max pooling layer

In this layer we loose even more definition and spatial reference. But we gain more in __feature isolation__. The images are much smaller (now `12x12`) but offer enough to distinguish those traits we mentioned before.

````python
# Visualize max pooling results for the 5th layer
maxpooling_f = K.function([model.layers[0].input], model.layers[4].output)
MP1 = maxpooling_f([X])
MP1 = np.squeeze(MP1)
print("MP1 shape : ", MP1.shape)

pl.figure(figsize=(12, 12))
pl.suptitle('maxpool')
nice_imshow(pl.gca(), make_mosaic(MP1, 6, 6), cmap=cm.binary)
````

####  Visualizing drop out

As indicated before I, initially, wasn't able to detect any visual differences between the previous layer and this one. This was due to using the validation phase rather than the training phase. So if we change the learning phase from validation to training mode, then we can actually see some fo those differences.

````python
# Visualize drop out results for the 6th layer in training mode
dropout1_f = K.function([model.layers[0].input, K.learning_phase()], model.layers[5].output)
# output in test mode = 0
# D1 = dropout1_f([X,0])[0]
# output in train mode = 1
D1 = dropout1_f([X,1])[0]
D1 = np.squeeze(D1)
print("D1 shape : ", D1.shape)

pl.figure(figsize=(12, 12))
pl.suptitle('dropout')
nice_imshow(pl.gca(), make_mosaic(D1, 6, 6), cmap=cm.binary)
````

#### Visualizing the final output

At last we can see a representation of the final layer ; this is probably not the best graph but it shows that we're now down to a single dimension array. There are 10 values, one for each of the categories, and the highlighted output neuron corresponds to the digit we were looking at, i.e. `7` (and `9` for the Z sample we used to show how the filters triggered different features in an image).

````python
# Visualize max pooling results for the 5th layer
softmax_f = K.function([model.layers[0].input, K.learning_phase()], model.layers[10].output)
S1 = softmax_f([X,0])[0]
print("S1 shape : ", S1.shape)
S1 = S1.reshape(S1.shape[0],1)
pl.figure()
pl.title('output for digit 7')
nice_imshow(pl.gca(), S1, vmin=0, vmax=1, cmap=cm.binary)
````
